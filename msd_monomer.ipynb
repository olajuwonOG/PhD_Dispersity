```python
#!/usr/bin/python

import sys, string
from numpy import *

nconf = 47
iskip = 0
dr = 0.2

file = 'MSD_xyz_nobinning'  # Output file for MSD

def MSD():
    msd = zeros(nconf, float32)  # Initialize an array to store MSD values

    for kconf in range(iskip + 1, nconf + iskip):
        print("Configuration:", kconf)

        # Read the configuration file
        sys.stdin.readline()
        sys.stdin.readline()
        sys.stdin.readline()

        line = sys.stdin.readline()
        fields = line.split()
        num = fields[0]

        sys.stdin.readline()
        line = sys.stdin.readline()
        [xm, xp] = map(float, line.split())
        line = sys.stdin.readline()
        [ym, yp] = map(float, line.split())
        line = sys.stdin.readline()
        [zm, zp] = map(float, line.split())

        xbox = xp - xm
        ybox = yp - ym
        zbox = zp - zm

        for j in range(1, dim):
            line = sys.stdin.readline()
            [ii, molj, typej, x1, x2, x3, n1, n2, n3] = string.split(line)
            k = int(ii)
            mol[k] = int(molj)
            xc[k] = xbox * (float(x1) - 0.5)
            yc[k] = ybox * (float(x2) - 0.5)
            zc[k] = zbox * (float(x3) - 0.5)

        # Calculate the MSD for the current configuration
        msd[kconf - iskip - 1] = compute_MSD()

    # Write the MSD values to the output file
    OUT = open(file, 'w')
    OUT.write("# Configuration MSD\n")
    for kconf in range(nconf):
        OUT.write("%4i %10.6f\n" % (kconf + iskip + 1, msd[kconf]))
    OUT.close()

def compute_MSD():
    squared_displacements = []

    # Loop through the particles and calculate squared displacements
    for i in range(1, dim):
        x0 = xc[i]
        y0 = yc[i]
        z0 = zc[i]

        x = xc[i]
        y = yc[i]
        z = zc[i]

        squared_displacement = (x - x0) ** 2 + (y - y0) ** 2 + (z - z0) ** 2
        squared_displacements.append(squared_displacement)

    # Calculate the mean squared displacement
    msd = sum(squared_displacements) / len(squared_displacements)
    return msd

# Initialize variables and arrays
natoms = 0
num1 = 0
num2 = 0
num2p = 0
dim = 0
xc = []
yc = []
zc = []
typea = []
mol = []
kcc = zeros(nconf + 1, float32)
MSD()

print("MSD calculations complete.")
```
