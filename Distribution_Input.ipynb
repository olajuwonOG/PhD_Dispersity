```python
"""For Schulz Zimm"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

def compute_values(z, Mn, lower_limit, upper_limit):
    x_values = np.linspace(lower_limit, upper_limit, 1000)
    prob = []
    for x in x_values:
        f = ((z**(z+1)/gamma(z))*(x**(z)/Mn**(z+1))*np.exp(-z*x/Mn))
        prob.append(f)
    return x_values, prob

z = 1.8
Mn = 25
lower_limit = 3
upper_limit = 80
x_values, prob = compute_values(z, Mn, lower_limit, upper_limit)

plt.plot(x_values, prob)
plt.xlabel('x')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import skewnorm
import matplotlib.pyplot as plt

# generate random numbers from a skew normal distribution
a = -33 # shape parameter
loc = 100 # location parameter
scale = 30 # scale parameter
samples = skewnorm.rvs(a, loc=loc, scale=scale, size=800)
monomers_per_polymer = samples
list_mon = list(monomers_per_polymer)
monomers_per_polymer = monomers_per_polymer.astype(int)
print("mon per pol", monomers_per_polymer.size)
z = Counter(list_mon)
data = pd.DataFrame({"No_chains":z})
print("This is the data ",data)
data.index
data.reset_index(inplace = True)
data = data.rename(columns = {'index': "No_of_monomers"})
data['weight_fraction'] = data["No_of_monomers"]*data["No_chains"]/sum(data["No_of_monomers"]* data["No_chains"])
data['number_fraction'] = data["No_chains"]/sum(data["No_chains"])
print("Print data",data)
Mw = sum((data['No_of_monomers']*data['weight_fraction']))
print("Average molecular weight is ", Mw)
Mn = sum((data['No_of_monomers']*data['No_chains']))/sum(data['No_chains'])
print("number average molecular weight is ", Mn)
D = Mw/Mn
print("Dispersity is ", D)
# plot a histogram of the generated samples
plt.hist(samples, bins=50, density=True, edgecolor = "black")
plt.title('Skewed Normal (Mw = 80),' '  D = ' +str(round(D, 2))+'', fontsize = 16, fontweight = 'bold', fontname="Arial")
plt.savefig('normal_distribution.png')
plt.show()
```
