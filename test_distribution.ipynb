```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rv_continuous
import numpy as np
import matplotlib.pyplot as plt
import random
from collections import Counter
import pandas as pd
from scipy.optimize import brentq
import math
class Schulz_Zimm(rv_continuous):
    def _pdf(self, x, z, Mn):
        self.x = x
        self.z = z
        self.Mn = Mn
        #return (z**(z+1)/np.math.gamma(z+1))*(x**(z-1)/Mn**(z))*np.exp(-z*x/Mn)
        #return (z**(z+1)/np.math.gamma(z+1))*(x**(z-1)/Mn**(z))*np.exp(-z*x/Mn)
        return ((z**(z+1)/np.math.gamma(z+1))*(x**(z)/Mn**(z+1))*np.exp(-z*x/Mn))

# Parameters for the distribution
np.random.seed(50)
z = 1  # Shape parameter
Mn = 17.9  # Scale parameter
num_samples = 1000
#b = 350
# Create an instance of the Schulz - Zimm distribution to generate random samples from the distribution
SZ = Schulz_Zimm(momtype=1, a = 0,  name='SZ') # The convergence of the root-finding algorithm can be sensitive to the choice of parameters.
monomers_per_polymer = SZ.rvs(z=z, Mn=Mn, size=num_samples)
#monomers_per_polymer = brentq(SZ.ppf, 0, 1000, args=(random.random(), z, Mn), maxiter=1000)
# Create a histogram to visualize the distribution
plt.hist(monomers_per_polymer, bins=15, edgecolor = "black", density=True)
# Calculate the theoretical distribution for visualization
x_vals1 = np.linspace(0, max(monomers_per_polymer), num_samples)
y_vals1 = SZ.pdf(x_vals1, z = z, Mn=Mn)
#plt.plot(x_vals, y_vals / np.trapz(y_vals, x_vals), color='red', label='Theoretical Distribution')
plt.plot(x_vals1, y_vals1 / np.trapz(y_vals1, x_vals1), color='red', label='Theoretical Distribution')
#plt.ylim([0, 0.07])
plt.xlabel('Chain Length')
plt.ylabel('Probability Density')
plt.title('Schulz Zimm Distribution')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rv_continuous
import numpy as np
import matplotlib.pyplot as plt
import random
from collections import Counter
import pandas as pd
from scipy.optimize import brentq
import math
class Schulz_Zimm(rv_continuous):
    def _pdf(self, x, z, Mn):
        self.x = x
        self.z = z
        self.Mn = Mn
        #return ((z**(z+1)/np.math.gamma(z+1))*(x**(z)/Mn**(z+1))*np.exp(-z*x/Mn)) #using the weight fraction formula
        return (z**(z+1)/np.math.gamma(z+1))*(x**(z-1)/Mn**(z))*np.exp(-z*x/Mn) #using the probability density formula
        #return ((z**(z+1)/gamma(z+1))*(x**(z)/Mn**(z+1))*np.exp(-z*x/Mn))

# Parameters for the distribution
np.random.seed(100)
z = 1  # Shape parameter
Mn = 200  # Scale parameter
num_samples = 1000
#b = 350
# Create an instance of the Schulz - Zimm distribution to generate random samples from the distribution
SZ = Schulz_Zimm(momtype=1, a = 0,  name='SZ') # The convergence of the root-finding algorithm can be sensitive to the choice of parameters.
monomers_per_polymer5 = SZ.rvs(z=z, Mn=Mn, size=num_samples)
#monomers_per_polymer = brentq(SZ.ppf, 0, 1000, args=(random.random(), z, Mn), maxiter=1000)
# Create a histogram to visualize the distribution
plt.hist(monomers_per_polymer5, bins=15, edgecolor = "black", density=True)
# Calculate the theoretical distribution for visualization
x_vals5 = np.linspace(0, max(monomers_per_polymer5), num_samples)
y_vals5 = SZ.pdf(x_vals5, z = z, Mn=Mn)
#plt.plot(x_vals, y_vals / np.trapz(y_vals, x_vals), color='red', label='Theoretical Distribution')
plt.plot(x_vals5, y_vals5 / np.trapz(y_vals5, x_vals5), color='red', label='Theoretical Distribution')
#plt.ylim([0, 0.07])
plt.xlabel('Chain Length')
plt.ylabel('Probability Density')
plt.title('Schulz Zimm Distribution')
plt.legend()
plt.show()
```

```python
#monomers_per_polymer = [math.ceil(x) for x in monomers_per_polymer]
monomers_per_polymer = monomers_per_polymer5
monomers_per_polymer = monomers_per_polymer.astype(int)
list_mon = list(monomers_per_polymer)

print("Number of the polymer chains", len(list_mon))
print(sum(monomers_per_polymer))
"""Calculating the dispersity"""
z = Counter(list_mon)
data = pd.DataFrame({"No_chains":z})
data.index
data.reset_index(inplace = True)
data = data.rename(columns = {'index': "No_of_monomers"})
data['weight_fraction'] = data["No_of_monomers"]*data["No_chains"]/sum(data["No_of_monomers"]* data["No_chains"])  
Mw = sum((data['No_of_monomers']*data['weight_fraction']))
print("Average molecular weight is ", Mw)
Mn = sum((data['No_of_monomers']*data['No_chains']))/sum(data['No_chains'])
print("number average molecular weight is ", Mn)
D = Mw/Mn
print("Dispersity is ", D)
a = np.array(list_mon)
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a, bins = 100, edgecolor = "black", density = True)
ax.axvline(a.mean(), color='r', linestyle='dashed', linewidth=3)
ax.axvline(Mw, color='g', linestyle='dashed', linewidth=3)
#plt.title('SZ (MW = , D = 41.4)', fontsize = 30, fontweight = 'bold', fontname="Arial")

plt.title(f'Schulz - Zimm Distribution, D = {math.ceil(D)}, Mw = {round(Mw,0)} & N = {num_samples}', fontsize = 18, fontweight = 'bold', fontname="Arial")
plt.xlabel("Number of Beads/Molecular Weights of chains", fontsize = 20, fontname="Arial")
plt.ylabel("Probability (Weight fraction)", fontsize = 20, fontname="Arial")
#plt.legend(loc='upper right')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
#plt.ylim([0, 0.0175])
#plt.xlim([0,1200])
plt.show()
```
