```python
import sys, string
import numpy as np
from math import pi, sqrt
from scipy.optimize import minimize_scalar

# INPUT PARAMETERS
nconf = 1  # number of configurations to average over
iskip = 0  # number of configurations to skip before taking data
numk = 25
numg1 = numk + 1

file = 'S_k_1.csv'

# Zero averages:
nk = np.zeros(numg1, int)
kcc = np.zeros(numg1, float)
S11 = np.zeros(numg1, float32)
S44 = np.zeros(numg1, float32)
SS11 = np.zeros(numg1, float32)
SS44 = np.zeros(numg1, float32)

# Initialize variables to collect data across configurations
natoms = 0
num1 = 0
num4 = 0

for iconf in range(nconf):
    # Read each configuration
    for i in range(iskip + 1):
        line = sys.stdin.readline()
        if not line:
            break

    if not line:
        break

    line = sys.stdin.readline()
    fields = line.split()
    natoms = int(fields[0])
    dim = natoms + 1
    xc = np.zeros(dim, float32)
    yc = np.zeros(dim, float32)
    zc = np.zeros(dim, float32)
    typea = [0] * dim
    mol = [0] * dim

    # Read box dimensions
    line = sys.stdin.readline()
    [xm, xp] = map(float, sys.stdin.readline().split())
    [ym, yp] = map(float, sys.stdin.readline().split())
    [zm, zp] = map(float, sys.stdin.readline().split())
    xbox = xp - xm
    dkx = 2 * pi / xbox
    Lx = numk * dkx
    ybox = yp - ym
    dky = 2 * pi / ybox
    Ly = numk * dky
    zbox = zp - zm
    dkz = 2 * pi / zbox
    Lz = numk * dkz
    vol = xbox * ybox * zbox

    for j in range(1, dim):
        line = sys.stdin.readline()
        [ii, molj, typej, q, x1, x2, x3, n1, n2, n3] = line.split()
        k = int(ii)
        typea[k] = int(typej)
        if iconf == 0:
            if typea[k] == 1:
                num1 += 1
            elif typea[k] == 4:
                num4 += 1
        mol[k] = int(molj)
        xc[k] = xbox * (float(x1) - 0.5)
        yc[k] = ybox * (float(x2) - 0.5)
        zc[k] = zbox * (float(x3) - 0.5)

    # Calculate structure factor for the current configuration
    for ll in range(0, numg1):
        for mm in range(0, numg1):
            for nn in range(0, numg1):
                s1c = s1s = s2c = s2s = s3c = s3s = s4c = s4s = 0
                kx = ll * dkx
                ky = mm * dky
                kz = nn * dkz
                kk = sqrt(kx * kx + ky * ky + kz * kz)
                jj = int(kk / dkx)
                if jj < numg1 and jj > 0:
                    kcc[jj] = kcc[jj] + kk
                    nk[jj] = nk[jj] + 1
                    for ii in range(1, dim):
                        kdotr = kx * xc[ii] + ky * yc[ii] + kz * zc[ii]
                        if typea[ii] == 4:
                            s4c = s4c + cos(kdotr)
                            s4s = s4s + sin(kdotr)
                        elif typea[ii] == 1:
                            s1c = s1c + cos(kdotr)
                            s1s = s1s + sin(kdotr)

                    S11[jj] += (s1c * s1c + s1s * s1s)
                    S44[jj] += (s4c * s4c + s4s * s4s)

# Calculate the average structure factor over all configurations
for k in range(1, numg1):
    kcc[k] = kcc[k] / nk[k]
    SS11[k] = S11[k] / (nk[k] * (num1))
    SS44[k] = S44[k] / (nk[k] * (num4))

# Write the results to a file
OUT = open(file, 'w')
OUT.write("k, S11, S44\n")
for ig in range(1, numg1):
    OUT.write("%8.4f, %8.4f, %8.4f\n" % (kcc[ig], SS11[ig], SS44[ig]))
OUT.close()
```
